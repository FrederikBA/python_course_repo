```python
import movies
import matplotlib.pyplot as plt

license_distribution = movies.license_distribution()

plt.pie(license_distribution.values(), labels = license_distribution.keys(), autopct='%1.1f%%')
plt.title('Movie License Distribution', fontsize=12)
plt.show()
```

```python
import movies
import matplotlib.pyplot as plt

pg_percentage_by_year = movies.percentage_of_pg_rated_movies()

plt.bar( pg_percentage_by_year.keys(), pg_percentage_by_year.values(), width=0.4)
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
plt.title('Percentage of PG-Rated movies from 2001 to 2015', fontsize=12)
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percentage of PG-rated movies', fontweight='bold')

plt.show()
```

```python
import movies
import matplotlib.pyplot as plt

average_revenue_dict = movies.average_revenue_by_genre()

sorted_dict = {k: v for k, v in sorted(average_revenue_dict.items(), key=lambda item: item[1])}

plt.bar(sorted_dict.keys(), sorted_dict.values(), width=0.4)
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
plt.title('Average World Revenue by Movie Genre', fontsize=12)
plt.xlabel('Movie Genre', fontweight='bold')
plt.ylabel('Average Revenue in $', fontweight='bold')

print(sorted_dict)
```
