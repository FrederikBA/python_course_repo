```python
import movies
import matplotlib.pyplot as plt

license_distribution = movies.license_distribution()

plt.pie(license_distribution.values(), labels = license_distribution.keys(), autopct='%1.1f%%')
plt.show()
```

```python
import movies
import matplotlib.pyplot as plt

pg_percentage_by_year = movies.percentage_of_pg_rated_movies()

plt.bar( pg_percentage_by_year.keys(), pg_percentage_by_year.values(), width=0.4)
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percentage of PG-rated movies', fontweight='bold')
plt.show()
```
